```python
import yfinance as yf
import matplotlib.pyplot as plt

tickers = [
    "RELIANCE.NS", "TCS.NS", "HDFCBANK.NS", "INFY.NS", "ICICIBANK.NS",
    "HINDUNILVR.NS", "SBIN.NS", "BHARTIARTL.NS", "ITC.NS", "BAJFINANCE.NS",
    "LICI.NS", "HCLTECH.NS", "LT.NS", "KOTAKBANK.NS", "ASIANPAINT.NS",
    "AXISBANK.NS", "SUNPHARMA.NS", "MARUTI.NS", "TITAN.NS", "ULTRACEMCO.NS",
    "ONGC.NS", "ADANIENT.NS", "NTPC.NS", "POWERGRID.NS", "WIPRO.NS",
    "JSWSTEEL.NS", "INDUSINDBK.NS", "TATAMOTORS.NS", "NESTLEIND.NS", "BAJAJFINSV.NS",
    "TATASTEEL.NS", "TECHM.NS", "HDFCLIFE.NS", "M&M.NS", "COALINDIA.NS",
    "GRASIM.NS", "CIPLA.NS", "BPCL.NS", "APOLLOHOSP.NS", "EICHERMOT.NS",
    "HINDALCO.NS", "DIVISLAB.NS", "DRREDDY.NS", "ADANIPORTS.NS", "BRITANNIA.NS",
    "HEROMOTOCO.NS", "SBILIFE.NS", "TATACONSUM.NS", "BAJAJ-AUTO.NS", "UPL.NS"
]

data = yf.download(tickers, start="2020-01-01", end="2025-01-30")['Close']

if data.empty:
    print("No data found for the given tickers and date range.")
    exit()

plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(data.index, data[ticker], label=ticker, alpha=0.7, linewidth=1)

plt.title("NIFTY 50 Stock Performance (2020-2025)")
plt.xlabel("Date")
plt.ylabel("Closing Price (INR)")
plt.legend(fontsize=6, loc="upper left", ncol=2)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```
